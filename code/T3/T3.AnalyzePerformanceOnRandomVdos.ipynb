```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#Print accuracy, precision, recall, and F1 score and make a function to do this for any assessed dataframe with 'isRelated' and 'result' columns
def print_metrics(assessed):
    # print confusion matrix
    print("Confusion Matrix:")
    print(pd.crosstab(assessed['isRelated'], assessed['result'], rownames=['Actual'], colnames=['Predicted']))

    print("\nClassification Report:")
    accuracy = accuracy_score(assessed['isRelated'], assessed['result'])
    precision = precision_score(assessed['isRelated'], assessed['result'])
    recall = recall_score(assessed['isRelated'], assessed['result'])
    f1 = f1_score(assessed['isRelated'], assessed['result'])
    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")
```

```python
# Write a function to print the confusion matrix and all scores for the assessed dataframe from the data_path only

def print_confusion_matrix(path):
    original = pd.read_json("input/200_random_videos_2302.json", orient="records")
    results = pd.read_csv(path)
    assessed = pd.merge(original, results, on="id", how="inner")
    assessed['result'] = assessed['result'].astype(int)
    #convert 'isRelated' to numeric, coercing errors to NaN, and then drop rows with NaN values
    assessed = assessed[pd.to_numeric(assessed['isRelated'], errors='coerce').notna()]
    assessed['isRelated'] = assessed['isRelated'].astype(int)
    
    print_metrics(assessed)
```

```python
print_confusion_matrix("data/0shot.csv")
```

```python
print_confusion_matrix("data/2shot.csv")
```
