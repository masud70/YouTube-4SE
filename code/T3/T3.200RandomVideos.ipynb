```python
import pandas as pd
```

```python
df = pd.read_json('../../data/videos.json')
df.shape
```

```python
# Select 200 random videos
random_videos = df.sample(n=200, random_state=42)

# Add a column for the index of the random videos from 1 to 200
random_videos['index'] = range(1, len(random_videos) + 1)

# Select id, title, and description columns
random_videos = random_videos[['index', 'id', 'title', 'description']]

# Add a new column as url with the YouTube video URL
random_videos['url'] = 'https://www.youtube.com/watch?v=' + random_videos['id']

# Add a new column as isRelated with a default value of True
random_videos['isRelated'] = True

# Save the random videos to a new JSON file
random_videos.to_json('../../data/random_videos.json', orient='records')
```
